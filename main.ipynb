```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
from tkinter import *
from tkinter import ttk
from IPython.display import HTML
```

```python
win =Tk()

win.title("News Reader")

# win.maxsize(width=900,height=300)
win.minsize(width=500,height=500)
win.geometry('1980x1080')
```

```python
def function():
    url="https://news.google.com/topstories?tab=rn&hl=en-IN&gl=IN&ceid=IN:en"
    # Step 1: Get the html
    r=requests.get(url)
    htmlContent=r.content;
    # Step 2: Parse the html
    soup=BeautifulSoup(htmlContent,'html.parser')
#     print(soup.prettify)
    news=soup.find_all('a',class_="DY5T1d RZIKme")
    length=len(news)
#     print(length)
    news_contents = []
    list_links = []
    # print(news)
    for i in range(length):
        # Getting the news headline
        event=news[i].get_text()
        news_contents.append(event)
         # Getting the link of the article
        link = news[i].get('href')
        link=link[1:]
        link="https://news.google.com/"+link
        list_links.append(link)
        
     #   ::::::::::::::::::::: inserting datafram into treeView::::::::::::::::::::::::;;   
    df=pd.DataFrame({
        "News Headlines": news_contents,"Artical Links": list_links
    })
    
#     print(df)
    
    HTML(df.to_html(render_links=True, escape=False))
    cols = list(df.columns)
    tree = ttk.Treeview(win,height=35)
    
    tree.place (x=200,y=50)
    tree["columns"] = cols
    for i in cols:
        tree.column(i, anchor="w",width=530) 
        tree.heading(i, text=i, anchor='w')

    for index, row in df.iterrows():
        tree.insert("",0,text=length-index,values=list(row))
    
    
    btn=Button(win,text="Save as CSV file",command=SaveIt(df),width=15)
    btn.place(x=80,y=350)
    
    btn=Button(win,text="Save as Excel file",command=SaveItxl(df),width=15)
    btn.place(x=80,y=400)
    return None
```
